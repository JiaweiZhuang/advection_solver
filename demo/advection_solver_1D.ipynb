```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

```python
from advection_solver.operator import upwind_tendency, vanleer_tendency
```

```python
nx = 100
Lx = 1
dx = Lx/nx
dt = 0.01
u = np.ones(nx) * 0.5
c0 = np.zeros(nx)
c0[int(nx*0.2):int(nx*0.4)] = 1.0
```

```python
plt.plot(c0)
plt.title('initial condition')
```

```python
nt = 500

c = c0.copy()
for _ in range(nt):
    c += upwind_tendency(c, u, dx, dt)
    
plt.plot(c)
plt.title('upwind scheme')
```

```python
c = c0.copy()
for _ in range(nt):
    c += vanleer_tendency(c, u, dx, dt)
    
plt.plot(c)
plt.title('Second-order flux-limited scheme (VanLeer)')
```
